## subway, bus 파생변수 추가

```python
bus = pd.read_csv('../data/bus_feature.csv')
subway = pd.read_csv('../data/subway_feature.csv')
dt = pd.read_csv('../data/train_v2.csv')
```

위도와 경도를 거리로 환산하는 함수를 정의하겠습니다.

```python
from math import radians, cos, dist

def GCS_to_meter(x1, y1, x2, y2):
    meter_y = 6371 * radians(1)
    meter_x = meter_y * cos(radians((y1+y2)/2))
    x = list(map(lambda x : x * meter_x, (x1 , x2)))
    y = list(map(lambda x : x * meter_y, (y1 , y2)))
    distance = dist((x[0], y[0]), (x[1], y[1]))
    return distance*1000
```

역세권(500M 이내)인 아파트를 구하겠습니다.

```python
apt_coordinate = dt.drop_duplicates(subset='주소')[['주소', '좌표X', '좌표Y']]
station_area = []
for apt_address, apt_x, apt_y in apt_coordinate[['주소', '좌표X', '좌표Y']].values:
    for sub_x, sub_y in subway[['좌표X', '좌표Y']].values:
        if GCS_to_meter(sub_x, sub_y, apt_x, apt_y) <= 500:
            station_area.append(apt_address)
            break

    
print(station_area)
```

```python
len(station_area)
```

역세권 여부를 0, 1로 구분하고 기존 데이터셋과 merge하겠습니다.

```python
dt_subway = pd.merge(dt, pd.DataFrame({'주소' : station_area, 'station_area' : [1] * len(station_area)}), on='주소', how='left')
dt_subway['station_area'] = dt_subway['station_area'].fillna(0)
```
